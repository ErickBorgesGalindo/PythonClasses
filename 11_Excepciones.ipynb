<h1><b>Excepciones</b></h2> </p>

Las excepciones son utilizadas para el manejo de errores. Cuando se trata de errores, en cuento menos tengamos mejor, sin embargo es cierto también que estos son inevitables, siempre estamos propensos a encontrar un error o "bug" dentro de nuestros programas.

Conociendo esto y tomando en cuenta que los errores son parte del día a día del desarrollador, para evitar que nuestra aplicación se cierre al ocurrir un error, o que la experiencia del usuario en nuestro sistema sea mala, vamos a hacer uso de las excepciones

```python
#Sintaxis
try:
  run this code
except:
  Execute this code when there is an error
else:
  Whitout errors? Run this code.
finally:
  Always run this code.
```

<b>Ejemplo del error</b>

```python
numberOne = 5
numberTwo = 1
numberTwo = '1'
```

```python
print(numberOne + numberTwo)
#El tener un error directo en consola significa que nuestro programa ha muerto
```

<b>Aplicando try - except</b>

```python
#Try - Except
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except: 
  #Se ejecuta si se produce una excepcion
  print('Se ha producido un error')
```

<b>Aplicando Try - Except - Else</b>

```python
#Try - Except - Else (Opcional)
numberOne = 1
numberTwo = 2

try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except: 
  print('Se ha producido un error')
else:
  #Se ejecuta si no se produce ninguna excepción
  print('La ejecución continúa correctamente')
```

<b>Aplicando Try - Except - Else - Finally </b>

```python
#Try - Except - Else (Opcional) - Finally (Opcional)
numberOne = 1
#numberTwo = 2
numberTwo = '2'

try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except: 
  print('Se ha producido un error')
else:
  print('La ejecución continúa correctamente')
finally:
  #Esto siempre se ejecuta 
  print('La ejecución sigue')
```

<b>Analizando tipos de errores</b>

```python
numberOne = 5
numberTwo = 1
numberTwo = '1'

print(numberOne + numberTwo)
```

```python
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except TypeError: 
  #Se ejecuta unicamente si el error es en el tipo de dato
  print('Se ha producido un error')
```

```python
#Solucionado el error
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except TypeError: 
  print('Hubo un error')
  print('Pero ya lo arreglé')
  fixOne = int(numberOne)
  fixTwo = int(numberTwo)
  print(fixOne + fixTwo)
```

```python
#Hay muchos tipos de error
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except ValueError: 
  print('Hubo un error')
```

```python
try:
  print(numberOne + numberTwo)
  print('No hay ningun error')
except TypeError: 
  print('Se ha producido un TypeError')
except ValueError:
  print('Hubo un error de valor')
```

Lo normal y correcto es que nosotros sepamos qué se está ejecutando en cada sección específica de nuestro código. Así podemos intuir el tipo de error que se pudiera cometer y solucionarlo con estos bloques de código.
