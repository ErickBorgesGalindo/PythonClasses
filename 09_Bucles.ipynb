<h2><b>While</b></h2>

```python
#Sintaxis
while condition:
  aqui va el codigo
```

<h2><b>FOR</b></h2>

Su función es similar a la utilizada en otros lenguajes de programación, sin embargo aquí en Python se ocupa principalmente para iterar sobre una secuencia (listas, tuplas, diccionarios, strings, etc)

```python
#Sintaxis
for iterator in lst:
    aquí va el código
```

```python
numeros = [0, 1, 2, 3, 4, 5]
for contador in numeros:
    print(contador)
```

```python
language = 'Python'
#for letter in language:
    #print(letter)


for i in range(len(language)):
    print(language[i])
```

```python
numbers = (0, 1, 2, 3, 4, 5)
for number in numbers:
    print(number)
```
