<b>Importando un módulo</b>

```python
#Módulo para trabajar con operaciones matemáticas
import math
print(math.pi)
print(math.sqrt(16))
print(math.pow(2,3))
print(math.floor(9.84))
```

```python
#Podemos importar funciones específicas de cada modulo
from math import pi, sqrt, pow
print(pi)
print(sqrt(16))
print(pow(2,3))
print(math.floor(9.84))
```

```python
#tambien podemos importar todas las funciones de jalón
from math import *
print(pi)
print(sqrt(16))
print(pow(2,3))
print(floor(9.84))
```

```python
#Tambien podemos renombrar una función
from math import pi as NumPI
print(NumPI)
```
