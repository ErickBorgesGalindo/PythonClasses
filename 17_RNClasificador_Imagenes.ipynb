```python
#El primer paso siempre es importar las librerias
import tensorflow as tf #A ti ya te conocemos
import tensorflow_datasets as tfds #Esta nos permite descargar el set de datos de manera simple
```

```python
#Descargar set de datos de Zalando
datos, metadatos = tfds.load('fashion_mnist', as_supervised=True, with_info=True)
#nombre del set, descargar info de los metadatos, descargar más info de metadatos.
#si SALEN LINEAS ROJAS NO HAY PROBELMA 
```

```python
#Imprimir los metadatos para ver que trae el set
metadatos
```

```python
#Obtenemos en variables separadas los datos de entrenamiento (60k) y pruebas (10k)
datos_entrenamiento, datos_pruebas = datos['train'], datos['test']
```

```python
#Etiquetas de las 10 categorias posibles
nombres_clases = metadatos.features['label'].names
```

```python
nombres_clases
```

```python
#Cuando trabajamos con redes neuronales queremos normalizar los datos, en este caso usaremos
#Funcion de normalizacion para los datos (Pasar de 0-255 a 0-1)
#Hace que la red aprenda mejor y mas rapido


def normalizar(imagenes, etiquetas):
  imagenes = tf.cast(imagenes, tf.float32)
  imagenes /= 255 #Aqui lo pasa de 0-255 a 0-1
  return imagenes, etiquetas

#Normalizar los datos de entrenamiento y pruebas con la funcion que hicimos
#Osea, pasamos todos los datos, tanto de entrenamiento como pruebas por la función de normalización.
datos_entrenamiento = datos_entrenamiento.map(normalizar)
datos_pruebas = datos_pruebas.map(normalizar)

#Agregar a cache (guarda datos en memoria en lugar de disco, hace el entrenamiento mas rapido)
datos_entrenamiento = datos_entrenamiento.cache()
datos_pruebas = datos_pruebas.cache()
```

```python
#Mostrar una imagen de los datos de pruebas, de momento mostremos la primera
for imagen, etiqueta in datos_entrenamiento.take(1):
  break
imagen = imagen.numpy().reshape((28,28)) #Redimensionar, cosas de tensores, lo veremos despues

import matplotlib.pyplot as plt

#Dibujar dibujar
plt.figure()
plt.imshow(imagen, cmap=plt.cm.binary)
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
#Dibujar más imagenes junto a su categoría
plt.figure(figsize=(10,10))
for i, (imagen, etiqueta) in enumerate(datos_entrenamiento.take(25)):
  imagen = imagen.numpy().reshape((28,28))
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(imagen, cmap=plt.cm.binary)
  plt.xlabel(nombres_clases[etiqueta])
plt.show()
```

```python
#Crear el modelo
#Tomamos las bases del modelo anterior, solo que en este caso nuestra capa de entrada
#sí se define de manera manual

modelo = tf.keras.Sequential([
  tf.keras.layers.Flatten(input_shape=(28,28,1)), #Definimos capa de entrada, y le decimos que nuestros datos
  #de entrada estan en una forma de 28x28px y en 1 canal (blanco y negro)

  #Agregamos dos capas ocultas con 50 neuronas cada una y todas estas con la función Relu
  tf.keras.layers.Dense(50, activation=tf.nn.relu),
  tf.keras.layers.Dense(50, activation=tf.nn.relu),

  #Finalmente nuestra capa de salidas con 10 neuronas (las categorías)
  #Cuando trabajamos con redes de clasificación le agregamos la función softmax en la capa de SALIDA
  #esto para asegurar que la capa de salida siempre nos de la que mas se acerque a la predicción.
  tf.keras.layers.Dense(10, activation=tf.nn.softmax) 
])
```

```python
#Compilar el modelo
#Nuevamente reutilizamos codigo de nuestra red anterior
modelo.compile(
    optimizer='adam', #Utilizamos ADAM como optimizador
    loss=tf.keras.losses.SparseCategoricalCrossentropy(), #Nuestra función de pérdida cambia, es especial para este tipo de redes
    metrics=['accuracy']
)
```

```python
#Separamos en variables los numeros de entrenamiento y pruebas
num_ej_entrenamiento = metadatos.splits["train"].num_examples
num_ej_pruebas = metadatos.splits["test"].num_examples
```

```python
#Verificamos que se asignen correctamente
print(num_ej_entrenamiento)
print(num_ej_pruebas)
```

```python
#Entrenar una red con 7 datos de entrada y 7 de salida es muy simple, pero ahora tenemos 70k de entrada
#Esto es mucho y nos tomaría mucho tiempo. Para esto haremos uso del trabajo en lotes
#Permite que entrenamientos con gran cantidad de datos se haga de manera mas eficiente
TAMANO_LOTE = 32

#Shuffle y repeat hacen que los datos esten mezclados de manera aleatoria para que la red
#no se vaya a aprender el orden de las cosas
datos_entrenamiento = datos_entrenamiento.repeat().shuffle(num_ej_entrenamiento).batch(TAMANO_LOTE)
datos_pruebas = datos_pruebas.batch(TAMANO_LOTE) #Aqui ifnoramos shuffle y repeat, porque no estamos entrenando.
```

```python
#Entrenar la red
import math
historial = modelo.fit(datos_entrenamiento, epochs=5, steps_per_epoch= math.ceil(num_ej_entrenamiento/TAMANO_LOTE))
#Para entrenar usamos la función fit donde especificamos
#Datos de entrenamiento, vueltas que queremos dar a los datos, steps_per_epoc (pero no la explicamos porque según va a desaparecer)

#Podemos ver que a diferencia de nuestra primera red se tarda más, y tiene sentido porque es un proceso más complejo
#En cada vuelta podemos ver la prección (accuracy), lo normal o bueno es que terminemos con precisión de 88% o superior
```

```python
#Nuevamente graficamos la perdida de información que tuvimos por cada vuelta (o epoch)
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
#Haremos una cuadricula que indique como la red va a analizar cada dato de prueba y lo guarde en una cuadricula
# Va a marcar cada cuadricula con colores, si fue correcta (azul) o incorrecta (roja)
import numpy as np

for imagenes_prueba, etiquetas_prueba in datos_pruebas.take(1):
  imagenes_prueba = imagenes_prueba.numpy()
  etiquetas_prueba = etiquetas_prueba.numpy()
  predicciones = modelo.predict(imagenes_prueba)
  
def graficar_imagen(i, arr_predicciones, etiquetas_reales, imagenes):
  arr_predicciones, etiqueta_real, img = arr_predicciones[i], etiquetas_reales[i], imagenes[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  
  plt.imshow(img[...,0], cmap=plt.cm.binary)

  etiqueta_prediccion = np.argmax(arr_predicciones)
  if etiqueta_prediccion == etiqueta_real:
    color = 'blue'
  else:
    color = 'red'
  
  plt.xlabel("{} {:2.0f}% ({})".format(nombres_clases[etiqueta_prediccion],
                                100*np.max(arr_predicciones),
                                nombres_clases[etiqueta_real]),
                                color=color)
  
def graficar_valor_arreglo(i, arr_predicciones, etiqueta_real):
  arr_predicciones, etiqueta_real = arr_predicciones[i], etiqueta_real[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  grafica = plt.bar(range(10), arr_predicciones, color="#777777")
  plt.ylim([0, 1]) 
  etiqueta_prediccion = np.argmax(arr_predicciones)
  
  grafica[etiqueta_prediccion].set_color('red')
  grafica[etiqueta_real].set_color('blue')
  
filas = 5
columnas = 5
num_imagenes = filas*columnas
plt.figure(figsize=(2*2*columnas, 2*filas))
for i in range(num_imagenes):
  plt.subplot(filas, 2*columnas, 2*i+1)
  graficar_imagen(i, predicciones, etiquetas_prueba, imagenes_prueba)
  plt.subplot(filas, 2*columnas, 2*i+2)
  graficar_valor_arreglo(i, predicciones, etiquetas_prueba)
```

```python
#Ahora vamos a probar por imagen suelta.
imagen = imagenes_prueba[4] #AL ser la variable imagenes_prueba solo tiene lo que se le puso en el bloque anterior heheh
imagen = np.array([imagen])
prediccion = modelo.predict(imagen)

print("Prediccion: " + nombres_clases[np.argmax(prediccion[0])])
```
