<b>Retornar valores de una función</b> </p>
Una función puede retornar valores , si no lo hace su valor final sera <b>None</b>

```python
print(adicion())
```

Para retornar un valor con una función nosotros usamos  la palabra return, seguida de la variable que vamos a retornar

```python
def adicion():
  a = 2
  b = 3
  total = a + b
  return total
```

```python
print(adicion())
```

<b><h2>Funciones con parámetros</h2></b> </p>
A las funciones podemos añadirles parámetros para que trabajen con valores ingresados por el usuario


```python
def adicion (a,b):
  total = a+b
  return total
```

```python
print(adicion(1,2))
```

<b>Tipos de valores a retornar</b>

```python
# String
def name(name):
  return name

name('Erick')
```

```python
#Booleanos
def is_pair(n):
  if n%2 == 0:
    return True
  return False

print(is_pair(8))
print(is_pair(5))
```

```python
#Listas
def find_pair(n):
  pairs = []
  for i in range(n+1):
    if i%2 == 0:
        pairs.append(i)
  return pairs

print(find_pair(10))
```
